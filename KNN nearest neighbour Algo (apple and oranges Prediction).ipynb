# K-Nearest neighbour

## Data prep

```python
total_fruits = apple_data.shape[0] + oranges_data.shape[0]
print(total_fruits)
```

```python
x = np.zeros((total_fruits,3))
print(x.shape)
```

```python
x[:900,:-1] = apple_data
x[900:,:-1] = oranges_data
x[900:,-1] = 1
x[:900,-1] = 0

print(x)
```

```python
x_values = x[:, :-1]
y_values = x[:,-1]
print(x_values.shape)
print(y_values.shape)
```

```python
test_fruit = np.array([4,11])
```

```python
plt.scatter(test_fruit[0],test_fruit[1], color = "red", marker= "*", label = "test_fruit")
plt.scatter(x_values[:,0],x_values[:,1], c=y_values) # c here is keyword which is refering to the 0 and 1 for apples and oranges
plt.legend()
plt.show()
```

## Building Algo

```python
def distance(p1, p2):
    return np.sum((p2 - p1)**2)**0.5                     # Distance formula for finding distance


p1 = np.array([1,2])
p2 = np.array([3,4])
```

```python
def knn(X, Y, test, k=5):   # k should always be small like 5, 7, 8 for accurate prediction
    m = X.shape[0]
    
    d = []
    for i in range(m):
        dist = distance(test , X[i])
        d.append((dist, Y[i]))
    
    d = np.array(sorted(d))[:,1]
    d = d[:k]
    t = np.unique(d,return_counts = True)
    idx = np.argmax(t[1])
    pred  = int(t[0][idx])
    
    if pred == 0:
        return "Apple"
    else:
        return "orange"
```
